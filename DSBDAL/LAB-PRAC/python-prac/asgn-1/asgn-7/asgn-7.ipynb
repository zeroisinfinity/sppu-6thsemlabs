```python
PARAGRAPH = """On April 6, 2026, India achieved a historic milestone in its pursuit of energy independence when the Prototype Fast Breeder Reactor (PFBR) at Kalpakkam, Tamil Nadu, successfully attained its first criticality, marking the official commencement of the second stage of the nation's ambitious three-stage nuclear power programme. This indigenous 500 MWe reactor, constructed by Bharatiya Nabhikiya Vidyut Nigam Limited (BHAVINI), represents a significant technological leap, making India only the second country after Russia to successfully commission a commercial-scale fast breeder reactor. The transition to Stage II is critical because Fast Breeder Reactors (FBRs) are designed to generate more fissile material than they consume; specifically, the PFBR utilizes a mixed oxide (MOX) fuel containing plutonium-239 recovered from the spent fuel of Stage I Pressurised Heavy Water Reactors (PHWRs), effectively closing the fuel cycle. By surrounding the core with a blanket of uranium-238, the reactor breeds new plutonium through nuclear transmutation, acting as a vital bridge to the eventual third stage which aims to harness India's vast thorium reserves. Prime Minister Narendra Modi hailed the achievement as a "defining step" towards clean energy security, noting that this technology not only reduces reliance on imported uranium but also significantly minimizes high-level nuclear waste, aligning perfectly with India's goal of achieving net-zero emissions by 2070 while ensuring a sustainable, long-term baseload power supply for its growing economy.
"""
```

```python
from nltk.corpus import stopwords
words = [w.lower() for w in words]
swords = stopwords.words('english')
```

```python
clean_words = [w for w in words if w not in swords]
clean_words
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
print(ps.stem('running'))
print(ps.stem('achieving'))
print(ps.stem('historical'))
```

```python
stemmed_words = [ps.stem(st) for st in clean_words]
stemmed_words
```

```python
from nltk.stem import WordNetLemmatizer as wnlemm
wnl = wnlemm()
wnl.lemmatize('achieved',pos='v')
```

```python
import string
string.punctuation
```

```python
only_words = [w for w in clean_words if w not in string.punctuation]
only_words
```

```python
proper_nouns = ['india', 'april', 'russia', 'kalpakkam', 'tamil', 'nadu', 'narendra', 'modi', 'bharatiya', 'nabhikiya', 'vidyut', 'nigam']
acronyms = ['pfbr', 'bhavini', 'mwe', 'mox', 'fbrs', 'phwrs']
proper_words = [w.upper() if w in acronyms else (w.title() if w in proper_nouns else w) for w in only_words]
proper_words
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag
final_tags = pos_tag(proper_words)
final_tags
```

```python
lemm_words = []
pos = None

for word , tag in final_tags:
    if tag.startswith('N'):
        pos = 'n'
    elif tag.startswith('J'):
        pos = 'a'
    elif tag.startswith('R'):
        pos = 'r'
    elif tag.startswith('V'):
        pos = 'v'
        
    if pos:
        lemm_words.append(wnl.lemmatize(word,pos=pos))
    else:
        lemm_words.append(word)
    pos = None

print(' '.join(lemm_words))
    
```
