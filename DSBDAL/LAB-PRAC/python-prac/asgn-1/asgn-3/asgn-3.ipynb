```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
age_group = np.array([
"18-25","18-25","18-25","18-25","18-25","18-25","18-25","18-25",
"26-35","26-35","26-35","26-35","26-35","26-35","26-35","26-35",
"36-45","36-45","36-45","36-45","36-45","36-45","36-45","36-45",
"46-60","46-60","46-60","46-60","46-60","46-60","46-60","46-60",
"61-75","61-75","61-75","61-75","61-75","61-75",
"76-100","76-100","76-100","76-100"
])

income = np.array([
3.2,4.1,2.8,3.6,np.nan,4.5,3.9,2.5,
6.5,7.2,np.nan,8.0,6.9,7.5,9.1,6.2,
10.5,11.2,9.8,np.nan,12.1,10.9,11.5,13.0,
14.5,15.2,13.8,16.1,np.nan,15.4,14.9,16.5,
12.2,11.8,np.nan,10.5,12.9,11.4,
8.1,np.nan,7.5,6.8
])

data = dict()
data['age_grp'] = age_group
data['income'] = np.round(income * 100000,2)
```

```python
df = pd.DataFrame(data,index=range(0,42,1))
```

```python
df.index.name = 'Sr No.'
df
```

```python
df.isnull().sum()
```

```python
df.fillna({'income':df['income'].mean()},
          inplace = True)
df.isnull().sum()
```

```python
np.round(df['income'].agg(['mean','median','min','max','std']),2)
```

```python
grp = df.groupby('age_grp')
grp
```
