```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
sns.__version__
dataset = sns.load_dataset("iris")


dataset

dataset1 = sns.load_dataset("iris")
dataset.columns = dataset1.columns

dataset.head()


dataset.info()


dataset.shape
```

```python
sns.set_theme(style='whitegrid' , palette='Set1')

for col in dataset.select_dtypes(include=['float64']).columns:
    plt.figure(figsize=(20,10))
    sns.histplot(data = dataset , x = col , hue = 'species' , 
                 kde = True , multiple='stack')
    plt.title(f'Distribution of {col} across species')
    plt.show()
```

```python
sns.set_theme(style='whitegrid' , palette='Set1')

for col in dataset.select_dtypes(include=['float64']).columns:
    plt.figure(figsize=(20,10))
    sns.boxplot(data = dataset , x = 'species' , y = col , palette='Set1')
    plt.title(f'Boxplot of {col} across species')
    plt.show()
```
