```python
df_new['CHAS'] = df_new['CHAS'].astype('int64')
df_new['RAD'] = df_new['RAD'].astype('int64')
df_new['TAX'] = df_new['TAX'].astype('int64')
```

```python
skew_values = df_new.skew()
print(skew_values)
```

```python
normal_cols = []
skewed_cols = []

for col in df_new.select_dtypes(include=['float64', 'int64']).columns:
    skewness = df_new[col].skew()
    
    if abs(skewness) <= 1:
        normal_cols.append(col)
        # Z-Score Logic
        mean, std = df_new[col].mean(), df_new[col].std()
        outliers = df_new[(df_new[col] < mean - 3*std) | (df_new[col] > mean + 3*std)]
        print(f"{col}: Normal (Skew: {skewness:.2f}) -> {len(outliers)} outliers.")
    else:
        skewed_cols.append(col)
        # IQR Logic
        Q1, Q3 = df_new[col].quantile(0.25), df_new[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df_new[(df_new[col] < Q1 - 1.5*IQR) | (df_new[col] > Q3 + 1.5*IQR)]
        print(f"{col}: Skewed (Skew: {skewness:.2f}) -> {len(outliers)} outliers.")

print("\nLists created!")
print(f"Normal: {normal_cols}")
print(f"Skewed: {skewed_cols}")
```

```python
skewed_cols.remove('CHAS')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style to look professional
sns.set_theme(style="whitegrid")

# We use a loop to create a separate subplot for every column in your list
plt.figure(figsize=(15, 10))

for i, col in enumerate(normal_cols, 1):
    plt.subplot(3, 3, i) # Creates a 3x3 grid
    sns.kdeplot(df_new[col], fill=True, color="royalblue", bw_adjust=1)
    plt.title(f"Smooth Bell Curve: {col}")
    plt.xlabel("") # Cleans up the axis for the report

plt.tight_layout()
plt.show()
```

```python
def plot_bell():
    for col in normal_cols:
        plt.figure(figsize=(6, 4))
        sns.kdeplot(df_new[col], fill=True, color="green")
        plt.title(f"Distribution of {col}")
        plt.show() # This will pop out a new, separate image for every column
def plot_box():
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df_new[skewed_cols])
    plt.title("Outliers in Skewed Distribution Columns (Z-Score Method)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
plot_bell()
```

```python
# 1. Calculate Mean and Std for all Normal columns at once
means = df_new[normal_cols].mean()
stds = df_new[normal_cols].std()

# 2. Define the Z-score boundaries (3 Standard Deviations)
upper_n = means + 3 * stds
lower_n = means - 3 * stds

# 3. Vectorized Clipping (Capping)
# axis=1 ensures Pandas matches each column's limit to the right column
df_new[normal_cols] = df_new[normal_cols].clip(lower=lower_n, upper=upper_n, axis=1)
```

```python
plot_bell()
```

```python
from sklearn.preprocessing import StandardScaler

# Centers normal data around 0
scaler_std = StandardScaler()
df_new[normal_cols] = scaler_std.fit_transform(df_new[normal_cols])
```

```python
plot_bell()
```

```python
plot_box()
```

```python
# Calculate Q1, Q3, and IQR for all skewed columns at once
Q1 = df_new[skewed_cols].quantile(0.25)
Q3 = df_new[skewed_cols].quantile(0.75)
IQR = Q3 - Q1

upper_s = Q3 + 1.5 * IQR
lower_s = Q1 - 1.5 * IQR

# Clip and then Log Transform
df_new[skewed_cols] = df_new[skewed_cols].clip(lower=lower_s, upper=upper_s, axis=1)
df_new[skewed_cols] = np.log1p(df_new[skewed_cols])
```

```python
plot_box()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Squashes skewed data between 0 and 1
scaler_minmax = MinMaxScaler()
df_new[skewed_cols] = scaler_minmax.fit_transform(df_new[skewed_cols])
```

```python
plot_box()
```
