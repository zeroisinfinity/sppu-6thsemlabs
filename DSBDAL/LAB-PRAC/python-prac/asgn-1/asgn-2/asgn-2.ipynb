```python
data2 = data.copy()
```

```python
data2['Roll no'] = data2['Roll no'].str[-3:]
data2['Roll no'] = data2['Roll no'].astype('float64')
data2.info()
```

```python
data_ayls = data2.select_dtypes(include=['float64','int64'])
data_ayls
```

```python
data_ayls.drop('Mobile No.' , inplace=True,axis=1)
```

```python
data_ayls['age'] = pd.Series(np.random.randint(17,23,size=data_ayls.shape[0]))
data_ayls.age.head()
```

```python
sns.boxplot(y=data_ayls['First year:   Sem 1'])
plt.show()

sns.boxplot(y=data_ayls['First year:   Sem 2'])
plt.show()

sns.boxplot(y=data_ayls["Second year:   Sem 1"])
plt.show()

sns.boxplot(y=data_ayls["Second year:   Sem 2"])
plt.show()

for col in data_ayls.columns:
        plt.figure(figsize=(6, 4))
        sns.kdeplot(data_ayls[col], fill=True, color="green")
        plt.title(f"Distribution of {col}")
        plt.show()
```

```python
skew_cols = []
norm_cols = []

for col in data_ayls.columns:

    if col in ['Roll no']:
        continue
    skewness = data_ayls[col].skew()
    outliers_n , outliers_s = pd.DataFrame() , pd.DataFrame()
    if abs(skewness) < 1:
        norm_cols.append(col)
        mean = data_ayls[col].astype('float64').mean()
        std = data_ayls[col].astype('float64').std()
        upper_l = mean + (3*std)
        lower_l = mean - (3*std)
        outliers_n[col] = data_ayls[col][(data_ayls[col] < lower_l) | (data_ayls[col] > upper_l)] 
        print(f'Outliers in {col} are --> {len(outliers_n[col])}\n')
        data_ayls.drop(outliers_n.index,inplace=True)
        data_ayls.reset_index(drop=True,inplace=True)

    else:
        skew_cols.append(col)
        q1 = data_ayls[col].quantile(0.25)
        q3 = data_ayls[col].quantile(0.75)
        iqr = q3 - q1
        upper = q3 + (1.5*iqr)
        lower = q1 - (1.5*iqr)
        outliers_s[col] = data_ayls[col][(data_ayls[col] < lower) | (data_ayls[col] > upper)] 
        print(f'Outliers in {col} are --> {len(outliers_s[col])}\n')
        data_ayls.drop(outliers_s.index,inplace=True)
        data_ayls.reset_index(drop=True,inplace=True)
   
```

```python
sns.boxplot(y=data_ayls['First year:   Sem 1'])
plt.show()

sns.boxplot(y=data_ayls['First year:   Sem 2'])
plt.show()

sns.boxplot(y=data_ayls["Second year:   Sem 1"])
plt.show()

sns.boxplot(y=data_ayls["Second year:   Sem 2"])
plt.show()

for col in skew_cols+norm_cols:
        plt.figure(figsize=(6, 4))
        sns.kdeplot(data_ayls[col], fill=True, color="green")
        plt.title(f"Distribution of {col}")
        plt.show()
```

```python
from sklearn.preprocessing import StandardScaler as mnmxsc
mnmx_scl = mnmxsc()
data_ayls[['age']+skew_cols+norm_cols] = mnmx_scl.fit_transform(data_ayls[['age']+norm_cols+skew_cols])
data_ayls.head()
```
