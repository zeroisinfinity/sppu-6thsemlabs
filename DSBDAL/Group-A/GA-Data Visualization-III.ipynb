```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
sns.__version__
dataset = sns.load_dataset("iris")


dataset

dataset1 = sns.load_dataset("iris")
dataset.columns = dataset1.columns

dataset.head()


dataset.info()


dataset.shape
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(20,10))
sns.histplot(data=dataset,x='sepal_length',hue='species'
             ,kde=True,palette='Set1',multiple='stack')
plt.title('Sepal len')
plt.show()
```
