```python
data = dict()
def create_ndata(key, float=False,suffix=str()):
    if not float:
            size = NUM_ROWS
            data[key+suffix] = np.asarray([
            np.random.randint(*RANGES[key],
                                size = NUM_ROWS,
                            )
            ])
    else:
            size = NUM_ROWS
            data[key+suffix] = np.asarray([
            np.round(np.random.uniform(*RANGES[key],
                                size = NUM_ROWS,
                            )
                     ,5)
            ])





def create_sdata(key, suffix=str(), weights=None, anchor=dict(), corresp=None):
    size = NUM_ROWS
    if (not corresp) or (not anchor):
        data[key+suffix] = np.asarray([
                                np.random.choice(RANGES[key],
                                size = NUM_ROWS,
                                p = weights
                                )
                            ])
    else:
        data[key+suffix] = np.asarray([
                                np.random.choice(RANGES[key],
                                size = NUM_ROWS
                                )
                            ])

        #yn = {k : v for k,v in zip([ele for ele in range(0,size) if data[key+suffix][ele] in RANGES[v]],list(anchor.keys()))}
        yn = dict()
        for ele in RANGES[key+suffix]:
            for k in anchor:
                if ele in RANGES[k]:
                    yn[ele] = k

        data[corresp] = np.asarray([
                                np.random.choice(RANGES[corresp],
                                size = 1,
                                p = anchor[yn[col]]
                                )
                                for col in data[key+suffix].flatten()
                            ]).reshape(1,-1)



nkeys = [key for key in RANGES.keys() if type(RANGES[key][0]) is not str ]
skeys = [key for key in RANGES.keys() if type(RANGES[key][0]) is str ]
```

```python
import numpy as np

df['GRADE'] = np.where((df['TOTAL_MARKS'] >= 0) & (df['TOTAL_MARKS'] < 525), 'F',
              np.where((df['TOTAL_MARKS'] >= 525) & (df['TOTAL_MARKS'] < 625), 'P',
              np.where((df['TOTAL_MARKS'] >= 625) & (df['TOTAL_MARKS'] < 725), 'C',
              np.where((df['TOTAL_MARKS'] >= 725) & (df['TOTAL_MARKS'] < 825), 'B',
              np.where((df['TOTAL_MARKS'] >= 825) & (df['TOTAL_MARKS'] < 925), 'B+',
              np.where((df['TOTAL_MARKS'] >= 925) & (df['TOTAL_MARKS'] < 1025), 'A',
              np.where((df['TOTAL_MARKS'] >= 1025) & (df['TOTAL_MARKS'] < 1125), 'A+',
              np.where(df['TOTAL_MARKS'] >= 1125, 'O', 'F'))))))))
df['GRADE']
```

```python

# better way

bins = [0, 300, 625, 725, 825, 925, 1025, 1125, 1500]
labels = ['F', 'P', 'C', 'B', 'B+', 'A', 'A+', 'O']

df['GRADE'] = pd.cut(df['TOTAL_MARKS'],bins=bins,labels=labels,include_lowest=True)

df['GRADE']
```
