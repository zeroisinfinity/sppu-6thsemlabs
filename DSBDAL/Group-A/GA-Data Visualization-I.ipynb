```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,8))
sns.countplot(data=dataset,x='survived',hue='sex',palette='Set1')
plt.title('Survivors categorized by gender')
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(20, 10))
sns.countplot(data=dataset, x='fare', palette='Set1')
plt.title('fare for each passenger is distributed')
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(20,10))
sns.histplot(data=dataset,x='fare',hue='sex',
             palette='Set1',multiple='stack',kde=True)
plt.title('fare for each passenger is distributed')
plt.show()
```
