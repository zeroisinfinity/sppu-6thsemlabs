```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings


warnings.filterwarnings('ignore')
sns.__version__
dataset = sns.load_dataset("titanic")
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(20,10))
sns.boxplot(data=dataset,x='sex',y='age' , hue='survived', palette='Set1')
plt.title('Survivors categorized by gender and plotted against age')
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(20,10))
sns.pairplot(data=dataset,palette='Set1',
             kind='kde',diag_kind='auto',
             hue = 'sex')

plt.title('Pairplot for titanic')
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(20,10))
sns.distplot(dataset['fare'])
plt.title('fare for each passenger is distributed')
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(20,10))
sns.distplot(dataset['fare'],kde=False)
plt.title('fare for each passenger is distributed')
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(50,10))
sns.jointplot(x='age',y='fare',data=dataset,
              palette='Set1',hue='sex',kind='scatter',
              height=10)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(50,10))
sns.jointplot(x='age',y='fare',data=dataset,
              palette='Set1',kind='hex',
              height=10)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(50,10))
sns.jointplot(x='age',y='fare',data=dataset,
              palette='Set1',kind='resid',
              height=10)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(100,100))
sns.rugplot(data=dataset['fare'])
plt.show()
```
