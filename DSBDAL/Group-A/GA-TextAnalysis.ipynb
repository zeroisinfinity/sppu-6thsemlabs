###  NLTK Tokenization: Sentence vs. Word

These imports allow you to break down unstructured text into manageable pieces (tokens). **Sentence** tokenization finds the boundaries of thoughts, while **Word** tokenization identifies individual units of meaning.

### 📋 Comparison & Logic Breakdown

$$
\begin{array}{|l|l|l|}
\hline
\textbf{Function} & \textbf{Level} & \textbf{Internal Logic} \\ \hline
\texttt{sent\_tokenize} & \text{Macro} & \text{Uses the \textbf{Punkt} model to split text into full sentences.} \\ \hline
\texttt{word\_tokenize} & \text{Micro} & \text{Splits sentences into words, punctuation, and symbols.} \\ \hline
\end{array}
$$

### 🧠 Why these are "Smart" Tools

Both functions are superior to standard Python string splitting because they understand linguistic context:

*   **Handling Ambiguity:**
    In the phrase *"The PFBR reactor (500 MWe) is in India."*:
    *   `sent_tokenize` knows the period after **MWe** isn't the end of the paragraph.
    *   `word_tokenize` knows **(500 MWe)** contains parentheses that should be separated from the numbers.

*   **Punctuation Treatment:**
    Unlike basic splitting, `word_tokenize` treats commas, brackets, and quotes as separate tokens, which is essential for later steps like **Part-of-Speech tagging**.

$$
\begin{array}{|l|l|}
\hline
\textbf{Input Type} & \textbf{Output Type} \\ \hline
\text{Raw Paragraph (String)} & \text{List of Sentences (Strings)} \\ \hline
\text{Single Sentence (String)} & \text{List of Words/Symbols (Strings)} \\ \hline
\end{array}
$$

### 🪓 Word Stemming (Porter Stemmer)

This code initializes the **Porter Stemmer**, one of the oldest and most common algorithms for stripping words down to their "stems" by chopping off suffixes.

### 📋 Technical Logic

$$
\begin{array}{|l|l|}
\hline
\textbf{Component} & \textbf{Linguistic Purpose} \\ \hline
\texttt{PorterStemmer()} & \text{An algorithmic "chopper" that uses heuristic rules to remove endings.} \\ \hline
\texttt{.stem('working')} & \text{The process of removing '-ing' to find the base form: \textbf{"work"}.} \\ \hline
\end{array}
$$

### 🧠 How it Works: The Rule-Based Approach

Stemming is **aggressive** and follow strict rules rather than a dictionary. Its goal is to group related words together even if the resulting "stem" isn't a real dictionary word.

$$
\begin{array}{|l|l|l|}
\hline
\textbf{Input Word} & \textbf{Stemmed Output} & \textbf{Rule Applied} \\ \hline
\text{"working"} & \text{"work"} & \text{Remove '-ing'} \\ \hline
\text{"achieved"} & \text{"achiev"} & \text{Remove '-ed' (Note: missing 'e')} \\ \hline
\text{"nuclear"} & \text{"nuclear"} & \text{No rule matches; remains same.} \\ \hline
\end{array}
$$

### ⚠️ Stemming vs. Lemmatization
Unlike **Lemmatization** (which uses a dictionary to get "achieve"), **Stemming** is faster but "crude." It might result in truncated words like **"achiev"** because it doesn't understand language—it only understands patterns.


````python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
ps.stem('running')
````

### 🔍 Analysis of Stemmed Output

The output you provided shows the result of applying the **Porter Stemmer** to your nuclear achievement paragraph. Notice how the algorithm aggressively "chops" words to find their base.

### 📋 Stemming Transformation Table

$$
\begin{array}{|l|l|l|}
\hline
\textbf{Original Word} & \textbf{Stemmed Result} & \textbf{Linguistic Observation} \\ \hline
\text{achieved / achievement} & \text{achiev} & \text{Suffix '-ed'/'-ment' removed; result is non-dictionary.} \\ \hline
\text{historical} & \text{histor} & \text{Suffix '-ical' chopped off.} \\ \hline
\text{success / successfully} & \text{success} & \text{Reduced to the common root for both variations.} \\ \hline
\text{ambitious} & \text{ambiti} & \text{Suffix '-ous' removed; end vowel changed.} \\ \hline
\end{array}
$$

### 🧠 Key Characteristics of this Output

1.  **Over-Stemming:** Words like **"achiev"** and **"ambiti"** are not real English words. This happens because the Porter Stemmer uses math-like rules rather than a dictionary.
2.  **Normalization:** Even though the spelling is "broken," both "achieved" and "achievement" became **"achiev"**. This is good for a computer because it can now count them as the same topic.
3.  **Noise Retention:** Notice that symbols like **"(", ")", ",",** and **"6"** remain unchanged. Stemmers only target the alphabetical suffixes of words.
4.  **Consistency:** The stemmer successfully grouped **"reactor"** and **"reactors"** (if present) under the same token, allowing for accurate frequency analysis of the nuclear topic.

### ⚠️ The "Crude" Nature
While **Stemming** is fast, it is less precise than **Lemmatization**. If you wanted "achiev" to remain "achieve," you would need the **WordNetLemmatizer** which uses a dictionary to keep words readable.

### ✅ The Fix: WordNet Lemmatization

By switching to `WordNetLemmatizer` and providing the **Part-of-Speech (POS)** tag, you have solved the "over-chopping" problem caused by the Stemmer.

### 📋 Technical Logic of the Fix

The `pos` parameter tells the lemmatizer exactly how the word is being used, allowing it to perform a dictionary-accurate transformation rather than a blind guess.

$$
\begin{array}{|l|l|l|l|}
\hline
\textbf{Input} & \textbf{POS Tag} & \textbf{Output} & \textbf{Logic Applied} \\ \hline
\text{'achieved'} & \text{'v' (Verb)} & \text{'achieve'} & \text{Converts past tense to base dictionary form.} \\ \hline
\text{'better'} & \text{'a' (Adjective)} & \text{'good'} & \text{Identifies the root quality (highly advanced).} \\ \hline
\text{'running'} & \text{'v' (Verb)} & \text{'run'} & \text{Strips the continuous tense perfectly.} \\ \hline
\text{'reactors'} & \text{'n' (Noun)} & \text{'reactor'} & \text{Removes the plural 's'.} \\ \hline
\end{array}
$$

### 🧠 Why this is "Fixed"
1. **Readable Output:** Unlike the Stemmer's `achiev`, the Lemmatizer returns `achieve`. This is essential if you need to present your analysis to humans.
2. **Grammar Awareness:** By using `pos='v'`, you are explicitly telling the machine: *"Treat this as an action."* This prevents the code from getting confused by words that could be both nouns and verbs.

### 📋 POS Shortcut Reference
$$
\begin{array}{|c|l|}
\hline
\textbf{Tag} & \textbf{Meaning} \\ \hline
\text{'n'} & \text{Noun (Default)} \\ \hline
\text{'v'} & \text{Verb} \\ \hline
\text{'a'} & \text{Adjective} \\ \hline
\text{'r'} & \text{Adverb} \\ \hline
\end{array}
$$

### 📝 TF-IDF Vectorization Explained

This code converts your text data into a numerical matrix using the **TF-IDF** (Term Frequency-Inverse Document Frequency) algorithm. This allows machine learning models to understand which words are the most important in your paragraph.

### 📋 Code Breakdown & Logic

$$
\begin{array}{|l|l|}
\hline
\textbf{Component} & \textbf{Linguistic / Technical Purpose} \\ \hline
\texttt{TfidfVectorizer()} & \text{Initializes the model to calculate word importance scores.} \\ \hline
\texttt{fit\_transform(t)} & \text{Learns the vocabulary and converts the text into a matrix.} \\ \hline
\texttt{v.shape} & \text{An attribute that reveals the dimensions (Rows, Columns) of the data.} \\ \hline
\end{array}
$$

### 🧠 The Math Behind the Logic

The **TF-IDF** score is calculated using two main factors to ensure that "fluff" words are ignored while "technical" words are highlighted:

1.  **Term Frequency (TF):** How many times a word appears in a specific sentence.
    *   *Example:* If "Reactor" appears 3 times in one sentence, it gets a higher score for that row.
2.  **Inverse Document Frequency (IDF):** How rare a word is across the *entire* document.
    *   *Example:* If "India" appears in every single sentence, its importance score **drops** because it isn't a unique identifier for that specific sentence.

### 📋 Analyzing `v.shape`

When you run `print(v.shape)`, it returns two numbers:

$$
\begin{array}{|c|c|l|}
\hline
\textbf{Value} & \textbf{Represents} & \textbf{Description} \\ \hline
\text{1st Number} & \text{Rows} & \text{The number of individual sentences or text units processed.} \\ \hline
\text{2nd Number} & \text{Columns} & \text{The total number of unique words (vocabulary size) found.} \\ \hline
\end{array}
$$

### ⚠️ Common Pitfalls

*   **Attribute vs Method:** `.shape` is a **Property** (a fact about the data), not a function. Adding `()` will cause a `TypeError` because you cannot "call" a dimension.
*   **Input Format:** `fit_transform` expects a list of **Strings**. If your variable `t` is still a list of **Tuples** (from the POS tagging step), you must convert it back to text first.


````python
# Fixed Import
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
# Note: fit_transform expects a list of STRINGS (sentences), not tuples
v = tfidf.fit_transform(t)

# Note: .shape is an attribute, not a function (no parentheses)
print(v.shape)
````

In your output, you see **228 rows** and many **0.0** values with a few **1.0** values. Here is why:

2.  **The "1.0" Value:** In Row 3, the column `2026` has a `1.0`. This means that in the 3rd position of your text, the word was "2026".
3.  **Sparsity:** Every other column in that row is `0.0` because a single position in a sentence can only hold one word at a time.
